```python
import torch
from torch import nn
```

```python
#查看可用GPU的数量
torch.cuda.device_count()
```

```python
#默认时张量创建在CPU上
X=torch.tensor([1,2,3])
X.device
```

```python
#张量利用.cuda()将其放置GPU上
X=torch.rand(2,3).cuda()
X
```

```python
Y=torch.ones(2,3).cuda()
X+Y
```

```python
#神经网络放置在GPU上
Net=nn.Sequential(nn.Linear(3,1))
Net=Net.to(device=torch.device('cuda'))
Z=torch.tensor([100,200,300],device=torch.device('cuda'),dtype=torch.float32)
Net(Z)
```

```python
Net[0].weight.data.device
```
